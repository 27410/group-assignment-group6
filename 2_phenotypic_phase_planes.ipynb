# Phenotypic Phase Plane analysis

We want to examine the correlation between growth and the production of the artemisinin precursor; dihydroartemisinic acid.

```python
# Initialising the required packages
from cobra.io import read_sbml_model # Loading the model
from cobra.flux_analysis.phenotype_phase_plane import production_envelope, add_envelope # Phase plane analysis text needed

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
```

```python
# Loading iYO844_modified, which we call model for simplicity
model = read_sbml_model("data/iYO844_modified.xml")
model
```

```python
# Looking at the dynamic of production of dihydroartemisinic aldehyde (DHAA) against biomass formation reaction
# The biomass formation reaction, which can be seen above: BIOMASS_BS_10

ppp_DHAA_biomass = production_envelope(model,
                    reactions=[model.reactions.ALDH1_CYP71AV1],
                    objective=model.reactions.BIOMASS_BS_10)

# ppp_DHAA_biomass
```

```python
# Generating the plot for the phase plane of DHAA vs biomass
ppp_plot = ppp_DHAA_biomass.plot(x='ALDH1_CYP71AV1', y='flux_maximum')
ppp_plot.get_figure().savefig("figures/initial_ppp") # Saving the plot as an png file.
```
